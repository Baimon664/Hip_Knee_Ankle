```python
img = cv2.imread("../test_dark.jpg", cv2.IMREAD_GRAYSCALE)
img_prep = preprocess(img, region="high")[:600]

# ret, thresh = cv2.threshold(img_prep, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

sigma = 2
k = 2
s1 = cv2.GaussianBlur(img_prep, ksize=(0,0), sigmaX=k*sigma)
s2 = cv2.GaussianBlur(img_prep, ksize=(0,0), sigmaX=sigma)
img_prep = cv2.subtract(s2, s1)

edges = cv2.Canny(img_prep, 0, 40)
# ret, edges = cv2.threshold(img_prep, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(img_prep, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(edges, cmap="gray")
plt.show()
```
